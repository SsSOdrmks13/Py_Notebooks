```python
import pandas as pd
```

```python
mycarsDataset = {
    'Company' : ["BMW", "BMW", "BMW", "Audi", "Audi", "Audi", "Audi", "Mercedes", "Mercedes", "RollsRoyce", "RollsRoyce", "RollsRoyce", "Ferrari", "Ferrari"] ,
    'Model' : ["BMW X1", "BMW X3", "BMW X5", "Audi Q3", "Audi Q5", "Audi A6", "Audi S5 sportsback", "Merc benz", "Merc Maybach", "Cullinan", "Dawn", "Ghost", "Roma", "Purosangue"]
}
```

```python
mySet = pd.DataFrame(mycarsDataset)
```

```python
mySet
```

```python
mySet['Prices'] = ["45L", "65L", "81L", "34L", "45L", "62L", "89L", "75L", "1Cr", "12Cr", "15Cr", "30Cr", "3Cr", "7Cr"]
```

```python
mySet
```

# Filtering the data

```python
print(mySet.loc[mySet['Company'] == "Audi"])
```

```python
print(mySet.loc[(mySet['Company'] == "BMW") & (mySet['Prices'] < "70L")])
```

# Regex Filtering

```python
mySet.loc[mySet['Model'].str.contains('n')]
```

```python
mySet.loc[~mySet['Model'].str.contains('n')]
```

```python
import re

mySet.loc[mySet['Model'].str.contains('bmw', flags=re.I, regex=True)]  #flags=re.I basically ignores the case sensitivity
```

```python
# for checking whether substring present

mySet.loc[mySet['Company'].str.contains('er[a-z]*', flags=re.I, regex=True)]
```

```python
# String starting with given substring

mySet.loc[mySet['Company'].str.contains('^Au[a-z]*', flags=re.I, regex=True)]  # ^ is used to indicate start of line
```

# Conditional Changes

```python
mySet.loc[mySet['Company'] == "BMW", 'Company'] = "Hyundai"
```

```python
mySet
```

```python
mySet.loc[mySet['Company'] == "Hyundai", 'Company'] = "BMW"
```

```python
mySet
```

```python
mySet.loc[mySet['Prices'] > "50L", 'Class'] = "Standardized"
mySet.loc[mySet['Prices'] < "50L", 'Class'] = "Modernized"
```
